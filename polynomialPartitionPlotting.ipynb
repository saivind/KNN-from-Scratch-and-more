```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import pandas as pd

# Importing the dataset
dataset = pd.read_csv('polynomial200.csv',header=None)

#storing the data into different variables
X = dataset.iloc[0:, 0:1].values
y = dataset.iloc[:, 1].values

#split dataset into training and Test dataset
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)

#plotting the training and test data

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12,10))

ax1.scatter(X_train, y_train)
ax1.set_title('Training Data')

ax2.set_title('Test Data')
ax2.scatter(X_test, y_test)

plt.show()

```
