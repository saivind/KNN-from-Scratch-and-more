```python
import  pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import seaborn as sns

# Read dataset from local storage to mc with headers X1,X2,test
mc = pd.read_csv("C:/Users/saivi/Documents/Machine Learning/Assignment 1/A1_datasets/A1_datasets/microchips.csv", header = None)
#changing the given data into array
b = np.asarray(mc)
nRow, nCol = b.shape #get the rows and coloumns 
#assigning the columns to individiual arrays
x = b[:, 0:1]
y = b[:, 1:2]
z = b[:, 2:3]
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.set_xlabel('Test1') #x-axis
ax.set_ylabel('Test2') #y-axis
for i in range(nRow):
    if z[i] == 1:
        ax.scatter(x[i], y[i], color='b', s=80, marker= '+')
    else:
        ax.scatter(x[i], y[i], color='r')
    plt.legend(('OK','Fail'))
plt.show()
```
